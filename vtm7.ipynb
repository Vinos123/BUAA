```python
import pandas as pd 
import os
from openpyxl import Workbook,load_workbook
import re

path = r'D:/Tang/vtm7.xlsx'#这个是所有excel文件夹的路径
pathfile = 'D:/Tang'#这个比上面那个多了个/，修改的化改这两个就行
data_list=[]
Depth_list=[]
f=open(path,'rb')
data=pd.read_excel(f)

```

```python
x=data.shape
```

```python
length=x[0]
data_list=list(data.iloc[0:length-1,0])
```

```python
for data_temp in data_list:
    if(data_temp.find('Depth')!=-1):
        Depth_list.append(data_temp)

```

```python
filepath_write = 'D:/Tang/Depthoutput.xlsx'
wb = load_workbook(filepath_write)
ws1 = wb['Sheet1']
row_ind=1
col_ind=1
for Depth in Depth_list:
    ws1.cell(row=row_ind,column=col_ind).value=Depth
    row_ind=row_ind+1
wb.save(filepath_write)


```
